### How to use Jupyter
When a cell is in edit mode:

  Shortcut  | Description
----------- | -----------
Shift+Enter | Execute the cell and move to the next one
Tab         | Indent code or trigger autocomplete
Esc         | Switch to command mode

When a cell is in command mode:

  Shortcut  | Description
----------- | -----------
Shift+Enter | Execute the cell and move to the next one
Double-click| Switch to edit mode
Enter       | Switch to edit mode

Additional shortcuts in command mode:
  Shortcut  | Description
----------- | -----------
A           | Insert a cell above
B           | Insert a cell below
C           | Copy the current cell
V           | Paste the copied cell below
D D         | Delete the current cell
M           | Convert the cell to Markdown
Y           | Convert the cell to Python code

To reset all cells:
* Go to the top menu and select Kernel -> Restart & Clear Output
